### 7.b - Compute statistics by group

Compute the proportion of female vs. male survivors of the RMS Titanic.

```python
proportion_male_survivors = titanic[titanic["Sex"] == "male"]["Survived"].mean()
proportion_male_survivors
```

Now compute the proportion of female vs. male survivors of the RMS Titanic, *along with the **s**tandard **e**rror of the **m**ean*. The **bold** type should give you a hint about the name of the method to compute the standard error. To do this, you'll need to combine the `groupby()` and `agg()` methods!

```python
import numpy as np

def compute_sem(data):
    return np.std(data, ddof=0) / np.sqrt(len(data))

proportion_survivors_gender = titanic.groupby("Sex")["Survived"].agg([("Proportion", "mean"), ("Standard Error", compute_sem)])
proportion_survivors_gender
```

What does this tell you about gender roles when the RMS Titanic was sunk?

_The proportion of female survivors is notably higher than that of male survivors. This discrepancy suggests that gender played a crucial role in determining survival outcomes during the Titanic disaster, potentially reflecting societal norms and the prioritization of female passengers for lifeboats._

Compute the proportion of survivors by cabin class and their standard error.

```python
def compute_sem(data):
    return np.std(data, ddof=0) / np.sqrt(len(data))

proportion_survivors_class = titanic.groupby("Pclass")["Survived"].agg([("Proportion", "mean"), ("Standard Error", compute_sem)])
proportion_survivors_class
```

What does this tell you about socio-economic status when the RMS Titanic was sunk?

_The proportion of survivors on the RMS Titanic varied significantly across different cabin classes, with first-class passengers having the highest survival rate and third-class passengers the lowest. This pattern underscores the influence of socio-economic status on survival outcomes, with wealthier individuals generally experiencing higher chances of survival compared to those in lower socio-economic classes._
