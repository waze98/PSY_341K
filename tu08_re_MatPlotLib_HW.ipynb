1. Make a plot of a straight line. Use `linspace()` to create the *x* values and the formula of a straight line, *y = a + b*x*, to create the *y* values (use an *a* and *b* of your choosing). You can pretend x and y are anything you like (x = time, y = international piracy or whatever).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# straight line
x = np.linspace(0, 14)
y = x*2

plt.xlabel("time")
plt.ylabel("international piracy")

plot_1 = plt.plot(x,y)
```

2. Make some data that are straight line values from the same straight line relationship as in 1. plus random noise. Plot these data.

```python
# values from plot 1 with random noise
x_data = [2, 4, 6, 8, 10, 12, 14]
y_data = np.random.randn(len(x_data))

plot_2 = plt.plot(x_data,y_data)
```

3. Plot the straight line from 1. and the data from 2. on the same graph. Make sure to add the standard annotations, including a legend.

```python
# plotting the two previous plots
plt.plot(x, y, label = "plot 1")
plt.plot(x_data, y_data, label = "plot 2")

# annotations
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Plot 1 vs Plot 2')

# legend
plt.legend()
```

4. Tinker around with your plot (colors, symbols, marker sizes, etc.) until you have a plot you would be happy to use in a presentation. 

```python
# make the plots
plt.plot(x, y, label = 'plot 1')
plt.plot(x_data,y_data, label = 'plot 2', color = "sienna", marker = "o")

# annotations
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Plot 1 vs Plot 2')
plt.legend()
```

5. Make 500 ***normally*** distributed random numbers and make a histogram of them.

```python
# size and normally distributed random numbers
my_sizes = 500 * np.random.rand(50) 

# histogram
plt.hist(my_sizes, edgecolor = "black", bins = 15)
```

6. Make 500 ***uniformly*** distributed random numbers (use `...rand()` instead of `...randn()`) and make a histogram of them.

```python
# size and uniform distribution random numbers
# Generate 500 uniformly distributed random numbers between 0 and 1
uni_rand_nums = np.random.rand(500)

# histogram
plt.hist(uni_rand_nums, edgecolor = "black", color = "green", bins = 15)
```

7. Plot the histograms from 5. and 6. in the same axes to compare the two distributions. Tinker around with the `color = ` and `alpha = ` arguments to `plt.hist()` until you're happy with your figure. Don't forget the axis labels and a legend! 

```python
# plotting the histograms from 5 and 6 
plt.hist(my_sizes, edgecolor = "black", bins = 15, alpha = 0.5, label = "Normal Distribution")
plt.hist(uni_rand_nums, edgecolor = "black", color = "green", bins = 15, alpha = 0.5, label = "Uniform Distribution")

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histograms of Uniform and Normal Distributions')

plt.legend()
```

8. Make a figure with 3 subplots, the first containing the plot of the data with a straight line (from 3.), and the second and third containing each of the 2 histograms created in 5. and 6. Try a 3x1 and  1x3 layout and show your favorite.

```python
# Creating figure with 3 subplots, 1x3 layout
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plotting first subplot, straight line
axs[0].plot(x, y, color='blue', label='Data')
axs[0].set_title('Data with Straight Line')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('International Piracy')

# Plotting second subplot, normal distribution histogram
axs[1].hist(my_sizes, edgecolor='black', bins=15, alpha=0.5, color='green')
axs[1].set_title('Histogram of Normal Distribution')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Frequency')

# Plotting third subplot, uniform distribution histogram
axs[2].hist(uni_rand_nums, edgecolor='black', bins=15, alpha=0.5, color='orange')
axs[2].set_title('Histogram of Uniform Distribution')
axs[2].set_xlabel('Value')
axs[2].set_ylabel('Frequency')

# Adjusting layout
plt.tight_layout()
```
