Write a function that uses `seaborn` to visualize data as we need it!
We'll work on the penguins dataset and plot the numeric variables *except for* body mass.

```python
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset("penguins")
```


To flex both our plotting and function writing muscles, let's write a function to do some plotting! Your function should:

* take as input the penguins  data frame
* allow the user to choose between a strip, violin, or box plot
* set one of the above three be the default
* have a docstr so users can get help() on it
* produce the plot requested by the user (of course!)
* provide a meanignful help

Write function

```python
def plot_penguins(penguins, plot_type='strip'):
    """
    Plot numeric variables from the penguins dataset.
    
    Parameters:
    - penguins (DataFrame): The penguins dataset.
    - plot_type (str): Type of plot to create. Options are 'strip' (default), 'violin', or 'box'.
    
    Returns:
    - None
    """
    # extract numeric variables excluding body mass
    numeric_vars = penguins.select_dtypes(include='number').columns.tolist()
    numeric_vars.remove('body_mass_g')
    
    # check plot type & create plot accordingly
    if plot_type == 'strip':
        sns.stripplot(data=penguins[numeric_vars])
    elif plot_type == 'violin':
        sns.violinplot(data=penguins[numeric_vars])
    elif plot_type == 'box':
        sns.boxplot(data=penguins[numeric_vars])
    else:
        raise ValueError("Invalid plot type. Choose from 'strip', 'violin', or 'box'.")
    
    # plot
    plt.title(f"Penguin Dataset - {plot_type.capitalize()} Plot")
    plt.xlabel("Variables")
    plt.ylabel("Values")
    plt.xticks(rotation=45)
    plt.show()
```

Get help on function

```python
# get help function
help(plot_penguins)
```

run function

```python
# run function with default plot type (strip)
plot_penguins(penguins)

plot_type = input("Choose a plot: strip, violin, or box plot: ")
plot_penguins(penguins, plot_type=plot_type)
```
